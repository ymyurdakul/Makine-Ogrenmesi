```python
import scipy.stats as stats
tips = sns.load_dataset("tips")
j=sns.jointplot(tips["total_bill"],tips["tip"],kind="reg",data=tips)
j.annotate(stats.pearsonr)

tips.head()
```

```python
from scipy import stats
import numpy as np
x = tips["total_bill"]
y = tips["tip"]
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
#print(p_value)
tips.head()

iris=sns.load_dataset("iris")
#iris.pivot_table(columns="species",aggfunc=sum)
x=sns.jointplot(x=iris["sepal_length"],y=iris["petal_length"],hue="species"])
x.annotate(stats.pearsonr)
```
