```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
hit=pd.read_csv("Hitters.csv")
df=hit.copy()
df=df.dropna()
dms=pd.get_dummies(df[["League","Division","NewLeague"]])
y=df["Salary"]
x_=df.drop(["Salary","League","Division","NewLeague"],axis=1).astype("float")
x=pd.concat([x_,dms[["League_N","Division_W","NewLeague_N"]]],axis=1)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=2)
```

```python
from lightgbm import LGBMRegressor
```

```python
hit=pd.read_csv("Hitters.csv")
df=hit.copy()
df=df.dropna()
dms=pd.get_dummies(df[["League","Division","NewLeague"]])
y=df["Salary"]
x_=df.drop(["Salary","League","Division","NewLeague"],axis=1).astype("float")
x=pd.concat([x_,dms[["League_N","Division_W","NewLeague_N"]]],axis=1)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=2)
```

```python
lgm_model=LGBMRegressor()
```

```python
lgm_model.fit(x_train,y_train)
```

```python
y_pred=lgm_model.predict(x_test,num_iteration=lgm_model.best_iteration_)
```

```python
np.sqrt(mean_squared_error(y_pred,y_test))
```

```python
light_grid={"learning_rate":[0.01,0.1,0.5,1]
           ,"n_estimators":[20,40,100,200,500,1000],
           "max_depth":[1,2,3,4,5,6,7,8]}
```

```python
lgbm=LGBMRegressor()
lgbm_cv_model=GridSearchCV(lgbm,light_grid,cv=10,n_jobs=-1,verbose=2)
```

```python
lgbm_cv_model.fit(x_train,y_train)
```

```python
lgbm_cv_model.best_params_
```

```python
lg_final_model=LGBMRegressor(learning_rate= 0.01, max_depth= 2, n_estimators= 1000)
```

```python
lg_final_model.fit(x_train,y_train)
```

```python
y_pred=lg_final_model.predict(x_test)
```

```python
np.sqrt(mean_squared_error(y_pred,y_test))
```
