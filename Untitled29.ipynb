```python
%config IPCompleter.greedy=True
from matplotlib import pyplot as plt
import numpy as np
yıllar=[1990,1991,1992,1993,1994]
ekmek_fiyatlari=np.arange(5)
diesel_fiyatları=np.arange(5)**2

fig,ax=plt.subplots()

ax.plot(yıllar,ekmek_fiyatlari,label="Ekmek Fiyatı",color="g")
plt.locator_params(axis="x",integer=True)
ex=ax.twinx()
ex.plot(yıllar,diesel_fiyatları,"r",label="Diesel fiyatı")
ax.fill_between(yıllar,ekmek_fiyatlari,diesel_fiyatları)
ax.legend()
ex.legend()
```
