```python
from scipy.stats import norm
veri_normal=norm.rvs(size=10000,loc=0,scale=1)
import seaborn as sns
import scipy.stats as stats
```

```python
sns.distplot(veri_normal,bins=100)
```

```python
import numpy as np
a1=stats.norm(loc=10,scale=2)
a2=stats.norm(loc=20,scale=8)
a3=stats.norm(loc=30,scale=6)

k=np.arange(40)

import matplotlib.pyplot as plt
```

```python
plt.plot(k,a1.pdf(k),marker="s")
plt.plot(k,a2.pdf(k),marker="o")
plt.plot(k,a3.pdf(k),marker="*")
```
