```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

fig,ax=plt.subplots()
x=np.arange(15)
y=x**2
ax.plot(x,y)
ax.annotate(s="Local Minima",xy=(0,0),xytext=(0,50),arrowprops=dict(facecolor="black",shrink=0.05))
ax.annotate(s="Local Maxima",xy=(14,200),xytext=(4,125),arrowprops=dict(facecolor="red"))
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection='3d')
x=np.arange(500)
y=np.arange(500)
X,Y=np.meshgrid(x,y)
Z=X*Y
ax.contour3D(X,Y,Z,cmap="hot")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='hot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
%notebook 
```
