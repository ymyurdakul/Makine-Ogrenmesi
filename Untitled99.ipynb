Bağımsız iki Örneklem T Testi-AB Testi 

```python
import pandas as pd
A=pd.DataFrame([30,27,21,27,29,30,20,
                20,27,32,35,22,24,23
               ,25,27,23,27,23,25,21
                ,18,24,26,33,26,27,28
                ,19,25])
B=pd.DataFrame([37,39,31,31,34,38,30,36,29,28,38,28,
               37,37,30,32,31,31,27,
               32,33,33,33,31,32,33,26,32,33,29])
```

```python
C=pd.concat([A,B],axis=1)
C
```

```python
ab=C
```

```python
ab.columns=["A","B"]
```

```python
ab.head()
```

```python
import numpy as np
grup_a=np.arange(len(A))
grup_a=pd.DataFrame(grup_a)
grup_a[:]="a"
A=pd.concat([A,grup_a],axis=1)
```

```python
A=pd.concat([A,grup_a],axis=1)
```

```python
A.head()
```

```python
grup_b=np.arange(len(B))
grup_b=pd.DataFrame(grup_b)
grup_b[:]="b"
B=pd.concat([B,grup_b],axis=1)
```

```python
AB=pd.concat([A,B])
AB
```

```python
AB.columns=["Gelir","Grup"]
AB.head()
```

```python
import seaborn as sns
sns.boxplot(x="Grup",y="Gelir",data=AB)
```

```python
#Varsayım Kontrolü Varyans homojenliği ve normallik 
from scipy.stats import shapiro
```

```python
shapiro(A)
```

```python
shapiro(B)
```

```python
import scipy.stats as stats
```

```python
#Ho anlamlı bir farklılık yoktur der 
#Ho hipotezi reddedilir
#Ümitler boşa çıkmamış istatiksel oalrak %95 güvenirlilikle
#Bu şans eseri çıkmamaıştır
stats.ttest_ind(A,B,equal_var=True)
```

```python
#Ho anmlamlı bir farklılık yoktur der w
#Ama reddeilir fark vardır
stats.mannwhitneyu(A,B)
```
