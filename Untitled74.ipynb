```python
import pandas as pd
```

```python
from pandas.api.types import CategoricalDtype
```

```python
import seaborn as sns
```

```python
d=sns.load_dataset("diamonds")
```

```python
d.head()
```

```python
d.cut=d.cut.astype(CategoricalDtype(ordered=True))
d.color=d.color.astype(CategoricalDtype(ordered=True))
d.clarity=d.clarity.astype(CategoricalDtype(ordered=True))
```

```python
d.cut.unique()
d.cut.value_counts().plot.bar()
```

```python
cut_ordinal=["Fair","Good","Very Good","Ideal","Premium"]
```

```python
d.cut=d.cut.astype(CategoricalDtype(categories=cut_ordinal,ordered=True))
```

```python
d.cut.value_counts().plot.barh()
```

```python
d.color.value_counts().plot.barh()
import numpy as np
```

```python
sns.barplot(data=d,x="cut",y="price",estimator=np.mean)
```

```python
d.groupby("cut")["price"].mean()
```

```python
#sns.jointplot(data=d,x="x",y="price",kind="reg",)
np.corrcoef(d["x"],d["F"])
#d["F"]=d.eval("x * y")
d.describe()
```

```python
sns.distplot(a=d.carat,kde=False,bins=100)
```

```python
sns.FacetGrid(data=d,hue="cut",height=5).map(sns.kdeplot,"price",shade=True)
```

```python
sns.catplot(data=d,x="cut",y="price",hue="color",kind="point",estimator=min)
```

```python
f=sns.load_dataset("flights")
```

```python
f.head()
```

```python

sns.factorplot(x="month",y="passengers",data=f,estimator=sum,palette="hot")
plt.xticks(Rotation=90)
```

```python
a=f.pivot_table(columns="month",values="passengers",aggfunc=sum)
a
```

```python
df=f.pivot_table(index="month",columns="year",values="passengers",aggfunc=np.sum)
fig=plt.figure(figsize=[16,9])
ax=fig.add_axes([0,0,1,1])
sns.heatmap(df,annot=True,ax=ax,fmt="d",cmap="hot")
```
