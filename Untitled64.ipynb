```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
planets = sns.load_dataset('planets')
planets.head()
```

```python
sns.factorplot("year",data=planets,hue="method",kind="count",order=range(2001, 2005))
```

```python

a=planets["method"].value_counts()
sns.barplot(x=a.index,y=a.values)
plt.xticks(Rotation=90)
plt.xlabel("Method")
plt.ylabel("Adet")
```

```python
sns.pairplot(data=planets,palette="hot")
```

```python
sns.distplot(planets["year"])
```

```python
sns.factorplot("year",kind="count",data=planets)
plt.xticks(Rotation=90)
```

```python
sns.jointplot(data=planets,x="mass",y="distance",kind="reg",color="red")
```

```python
f=planets.pivot_table(index="method",values=["mass"],aggfunc=max)
f.plot()
```
