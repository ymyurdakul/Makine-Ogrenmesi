```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sui=pd.read_csv("who_suicide_statistics.csv")
```

```python
sui.rename(columns={"country":"Ulke","year":"Yıl","sex":"Cinsiyet","age":"Yas","suicides_no":"Intihar_Sayısı","population":"Nufus"},inplace=True)
```

<h3>'1979 ile  2016 Yılları Arasındaki Ulkelere Gore Toplam Intihar Sayıları'</h3>

```python
fig=plt.figure(figsize=[14,4])
title=str(sui["Yıl"].min())+" ile  "+str(sui["Yıl"].max())+" Ulkelere Gore Yılları Arasındaki Intihar Sayıları"
a=sui.groupby("Ulke")["Intihar_Sayısı"].sum()
a=a.sample(20)
ax=fig.add_axes([0,0,1,1])
ax.bar(a.index,a.values,align="center",color="red")
plt.xticks(Rotation=90)
for i in range(len(a)):
    ax.annotate(s=str(a.values[i]),xy=(a.index[i],a.values[i]))
```

<h2 style="color:green">Türkiyede Yıllara ve Cinsiyete Göre Intıhar Sayısı ve Nufus</h2>

```python
tr_sui=sui[sui["Ulke"]=="Turkey"]
tr_sui.groupby(by=["Yıl","Cinsiyet"]).sum()
```

<h1 style="color:red;">Nufus ve Intıhar Sayısı Arasındaki ilişkiyi gösteren plot</h1>

```python
tr_sui.groupby(by=["Yas","Cinsiyet"]).sum()
tr_sui["Intihar Oranı"]=tr_sui.eval("Intihar_Sayısı/(Nufus/100)")
tr_sui["Intihar Oranı"].apply(lambda x:np.round(x,4))
sns.jointplot(data=tr_sui,x="Nufus",y="Intihar_Sayısı",kind="reg",color="red")
np.corrcoef(tr_sui["Nufus"],tr_sui["Intihar_Sayısı"])
```

```python
sns.pairplot(data=tr_sui,vars=["Intihar Oranı","Intihar_Sayısı","Nufus"],hue="Cinsiyet",kind="reg")
```

<h1 style="color:blue;">Yıllara gore Türkiyede nufus artısı</h1>

```python
nufus=tr_sui.groupby("Yıl").sum()["Nufus"]
sns.barplot(x=nufus.index,y=nufus.values,)
```

<h2 style="color:red">Türkiyede yas aralıklarına göre ve cinsiyete göre toplam intihar sayısı</h2>

```python
tr_sui.pivot_table(columns="Yas",values="Intihar_Sayısı",index="Cinsiyet",aggfunc="sum").plot(kind="bar")
```

<h2 style="color:red;">Türkiyede Yıllara ve cinsiyete göre toplam intihar sayısı</h2>

```python
e=tr_sui.pivot_table(columns=["Cinsiyet"],index="Yıl",values="Intihar_Sayısı",aggfunc="sum")
e.plot(kind="bar")
```

<h2 style="color:yellow">Dünya genelinde yas aralığına gore Intıhar Sayıları</h2>

```python
e=sui.pivot_table(columns="Cinsiyet",index="Yas",values="Intihar_Sayısı",aggfunc="sum")
e.plot(kind="line")
```

<h2 style="color:red">TOPLAM INTIHAR SAYISININ ENCOK OLDUGU İLK 30 ULKE</h2>

```python
fig=plt.figure(figsize=[16,9])
ax=fig.add_axes([0,0,1,1])
mx=sui.pivot_table(index="Ulke",aggfunc="sum",values="Intihar_Sayısı").sort_values(by="Intihar_Sayısı",ascending=False)
mx_h=mx.head(30)
d=mx_h.stack()
sns.barplot(x=d.index,y=d.values,ax=ax,palette="hot")
plt.xticks(Rotation=90)
```

<h2 style="color:red">TOPLAM INTIHAR SAYISININ ENAZ OLDUGU İLK 30 ULKE</h2>

```python
fig=plt.figure(figsize=[16,9])
ax=fig.add_axes([0,0,1,1])
mx=sui.pivot_table(index="Ulke",aggfunc="sum",values="Intihar_Sayısı").sort_values(by="Intihar_Sayısı",ascending=True)
mx_h=mx.head(30)
d=mx_h.stack()
sns.barplot(x=d.index,y=d.values,ax=ax,palette="cool")
plt.xticks(Rotation=90)
```

<h2 style="color:red">USA de INTIHAR VE NUFUS SAYILARININ HEATMAPI</h2>

```python
f=sui["Ulke"].str.contains("USA")
abd_sui=sui[f]
h=abd_sui.pivot_table(index="Yas",values=["Intihar_Sayısı","Nufus"])
sns.heatmap(h)
```
