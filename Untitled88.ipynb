```python
d=sns.load_dataset("diamonds")
```

```python
d.table.plot.box()
```

```python
sns.boxplot(d.table)
```

```python
Q1=d.table.quantile(0.25)
Q3=d.table.quantile(0.75)
IQR=Q3-Q1
alt_sinir=Q1-1.5*IQR
ust_sinir=Q3+1.5*IQR
print(alt_sinir,ust_sinir)
```

```python
f=(d.table<alt_sinir)|(d.table>ust_sinir)
```

```python
cd=d[~f].copy()
```

```python
sns.boxplot(cd.table)
```
