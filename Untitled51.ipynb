```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
tips=sns.load_dataset("tips")
#by time tip values
plt.figure(figsize=[16,9])
plt.subplot(2,2,1)
sns.boxplot(x="time",y="tip",data=tips,palette="deep",hue="sex")
#Akşam yemeklerinde ortalama olarak daha fazla tip verilmiş
plt.subplot(2,2,2)
sns.boxplot(x="time",y="total_bill",data=tips,palette="muted",hue="sex")
plt.subplot(2,2,3)
sns.boxplot(x="day",y="total_bill",data=tips,palette="colorblind",hue="sex")
plt.subplot(2,2,4)
sns.boxplot(x="sex",y="total_bill",hue="smoker",data=tips)
plt.tight_layout()

#sns.swarmplot(x="sex",y="tip",hue="smoker",data=tips)
```
