```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

data = pd.read_csv("gp.csv")
data[data["Rating"].between(4,5)]
s=data["Rating"].dropna().value_counts()
plt.scatter(s.index,s.values)
plt.xlabel("Frekans")
plt.ylabel("Aralık")
data["Installs"]=data["Installs"].apply(lambda x:x.split("+")[0])
data["Installs"]=data["Installs"].apply(lambda x:x.replace(",","."))
data["Installs"]=data["Installs"].apply(lambda x:x.replace(".",""))
data.drop(index=10472,inplace=True)
data["Installs"]=data["Installs"].astype(float)
data.plot(subplots=True)
data["Installs"].unique()
```

```python
data.rename(columns={"App":"Uygulama",
                     "Category":"Kategori",
                     "Rating":"Reyting",
                     "Reviews":"İnceleme",
                    "Size":"Boyut",
                     "Installs":"Yüklemeler"
                    },inplace=True)
data.drop(labels=["Kategori"],axis=1)
data.sample(10,axis=0)
```

```python
data.nsmallest(n=5,columns=["Yüklemeler"])
```

```python
data.nlargest(n=10,columns=["Yüklemeler"]).sort_values(by="Yüklemeler")
data.head()
```

```python
f1=data["Kategori"]=="ART_AND_DESIGN"
f2=data["Reyting"].between(2.0,4.0)
data["Reyting"]=data["Reyting"].dropna()
data.where(f1).dropna()
data.where(f2).dropna()
data["İnceleme"]=data["İnceleme"].apply(func=lambda l:l+"  Milyon")
data
```
