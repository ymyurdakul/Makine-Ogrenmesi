```python
import pandas as pd
import numpy as np
d=pd.read_csv("deprem.csv")
d.head(100)
d.groupby("Region-Name")["Mag"].max()
d["Region-Name"].value_counts()
d.groupby('Mag').describe()
d.groupby('Mag').aggregate(['min', np.median, max])
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.groupby(["sex","class"])["survived"].sum()
titanic.pivot_table("survived",index="sex",columns="class")
age=pd.cut(titanic["age"],[1,18,80])
fare=pd.cut(titanic["fare"],[1,20,1000])
fare=pd.qcut(titanic["fare"],3)
titanic.pivot_table("survived",index=["sex",age,"embark_town",fare],columns="class")
#titanic.head()
```
