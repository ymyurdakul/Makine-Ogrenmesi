```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
deprem=pd.read_csv("earthq.csv")
deprem.head()
```

Columns ları değiştirme

```python
deprem.rename(columns={"Date-Time":"Tarih","Mag":"Çap","Type":"Tip","Latitude":"Enlem","Longitude":"Boylam","Depth":"Derinlik","Region-Name":"Bolge"},inplace=True)
deprem.head()
```

```python
deprem.info()
```

```python
deprem.describe()
```

```python
#Tarihlere göre deprem sayısı
a=deprem["Tarih"].value_counts().head(10)
sns.barplot(x=a.index,y=a.values)
plt.xticks(Rotation=90)
```

```python
#Derinliklere göre deprem sayısı
a=deprem["Derinlik"].value_counts().head()
a.sort_index(ascending=True).plot(kind="line")
a
```

```python
#Çaplarına göre deprem sayıları
a=deprem["Çap"].value_counts().sample(10)
sns.barplot(x=a.index,y=a.values)
a
```

```python
#Tiplerine göre deprem sayıları
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
a=deprem["Tip"].value_counts()
sns.barplot(x=a.index,y=a.values,ax=ax)
plt.xticks(Rotation=90)
```

```python
sns.countplot(data=deprem.sample(15),x="Çap",hue="Tip")
```

```python
ankara_deprem=deprem[deprem["Bolge"].str.contains("Ankara")]
sns.countplot(ankara_deprem["Çap"])
plt.xticks(Rotation=90)
```

```python
deprem=deprem.dropna()
deprem.plot()
```

```python
#deprem["Tarih"]=pd.to_datetime(deprem["Tarih"].values)
#deprem.set_index("Tarih",inplace=True)
deprem.groupby("Bolge")["Çap"].max().head(200).plot(color="yellow")
plt.xticks(Rotation=90)
```

```python
deprem["Derinlik"]=deprem["Derinlik"].str.replace("-","0").astype(int)
sns.pairplot(data=deprem)
```
