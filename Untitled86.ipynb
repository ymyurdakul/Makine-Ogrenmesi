Tek Örnekleme Oran Testi
Problem:Dönüşüm Oranı test edilmek isteniyor 
H0:p=0.125
H1:p!=0.125
500 kişi girmmiş 40 tıklamış
Ornek oranı=0,08

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
count=40
nobs=500
value=0.125
```

```python
proportions_ztest(count,nobs,value)
```

```python
#%95 olasılıkla value 0.125 den farklı
```

```python
40/500
```

```python
proportions_ztest(count,nobs,value,alternative="smaller")
```
