```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
wh=pd.read_csv("wh_2017.csv") 
wh.drop(labels=["Happiness.Rank","Whisker.high","Whisker.low","Generosity","Dystopia.Residual"],axis=1,inplace=True)
wh.head()
```

<h2 style="color:orange">TÜRKİYEDE 2015-2016-2017 YILLARI ARASINDA KİŞİ BASI DUSEN MILLI GELIR VE MUTLULUK ORANLARININ DEGISIM GRAFIGI</h2>

```python
wh_2015=pd.read_csv("wh_2015.csv")
wh_2016=pd.read_csv("wh_2016.csv")
wh_2017=pd.read_csv("wh_2017.csv")
f_2015=wh_2015["Country"]=="Turkey"
f_2016=wh_2016["Country"]=="Turkey"
f_2017=wh_2017["Country"]=="Turkey"

k_2015=wh_2015[f_2015]
k_2016=wh_2016[f_2016]
k_2017=wh_2017[f_2017]
yıllar=pd.to_datetime(["2015-01-01","2016-01-01","2017-01-01"])
mutlulukOranları=[k_2015["Happiness Score"].values[0],k_2016["Happiness Score"].values[0],k_2017["Happiness.Score"].values[0]]
gOranları=[k_2015["Economy (GDP per Capita)"].values[0],k_2016["Economy (GDP per Capita)"].values[0],k_2017["Economy..GDP.per.Capita."].values[0]]
```

```python
fig=plt.figure(figsize=[5,5])
ax=fig.add_axes([0,0,1,1])
ax.plot(yıllar,gOranları,label="Kişi Başı Düşen Milli Gelir")
plt.xticks(Rotation=90)
ex=ax.twinx()
ex.plot(yıllar,mutlulukOranları,color="red",label="Mutluluk Oranları")
ax.legend()
ex.legend(loc="upper right")
ax.set_xlabel("Yıllar")
ax.set_ylabel("Oranlar")
```
